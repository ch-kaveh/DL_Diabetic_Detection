But for undrestanding the data and having more information about this columns we decided to find these outlyer and check howmany percantage of these data have Diabets

So at the first step we find the outlyers

we find lower and upper area based on  IQR = Q3 - Q1 formulia

```python
Q1 = np.percentile(Data['Insulin'], 25)
Q3 = np.percentile(Data['Insulin'], 75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Lower limit for area: {lower}")
print(f"Upper limit for area: {upper}")
```

```python
def Outlyer_checker() :
    Outliers_df = pd.DataFrame(columns=['Index', 'Insulin','Outcome'])   
    counter = 0
    WithDiabet = 0
    WithoutDiabet = 0
    Outliers = []
    for i in Data['Insulin'].index:
        if Data['Insulin'][i] > upper or Data['Insulin'][i] < lower:
            d = {"Index": i, "Insulin": Data['Insulin'][i], "Outcome": Data['Outcome'][i]}
            df2 = pd.DataFrame(d, index={counter})
            counter+=1
            Outliers_df = pd.concat([Outliers_df, df2])
            Outliers.append(i)
            if Data['Outcome'][i] == 1:
                WithDiabet += 1
            else :
                WithoutDiabet += 1
    #Outliers_df
    Output_Label = ['Yes', 'No']
    Output_Count = [WithDiabet, WithoutDiabet]
    print('Number of outlyers With Diabets : ',WithDiabet )
    print('Number of outlyers Without Diabets : ', WithoutDiabet)
    return Output_Label, Output_Count, Outliers
```

```python
Output_Label ,Output_Count, Outliers = Outlyer_checker()
plt.figure(figsize=(8, 8))
plt.pie(Output_Count,labels=Output_Label)
plt.show()
```

here we find the outlyers and find the output for them to have a better view of the outlyers 

now we can see almost half of the outlyers have diabet and half of them don't have

so we can delete the outlyers and it does not affect our training

```python
def ECDF(data):
    n = len(data)
    x =  np.sort(data)
    y = np.arange(1,n+1) / n
    return x,y
```

Now we preaper data and show the relation between the columns and find some statistic information which help us to find models for our data

```python
scaler = preprocessing.MinMaxScaler()
names = Data.columns
d = scaler.fit_transform(Data)
scaled_Data = pd.DataFrame(d, columns=names)
scaled_Data.head()
```

```python
X = scaled_Data.iloc[:,0:8]
Y = scaled_Data.iloc[:,-1]
```

# Shuffle data

```python
X_per = X.copy()
X_per.drop(X_per.index, inplace=True)
ftr = X.copy()
ftr.drop(ftr.index, inplace=True)

#Y_per = Y.copy() 
#Y_per.drop(Y_per.index, inplace=True)
#Y_per = pd.DataFrame(Y_per)

#lbl = Y.copy()
#lbl.drop(lbl.index, inplace=True)
#lbl = pd.DataFrame(lbl)
```

```python
list_per = np.random.permutation(len(X))
Y_per = []
for i in range(len(list_per)):
    idx_per = list_per[i]
    ftr = X.iloc[idx_per:idx_per+1,0:8]
    lbl = Y.iloc[idx_per]
    X_per = pd.concat([X_per, ftr])
    Y_per.append(lbl)
    
Y_per = pd.DataFrame(Y_per)
Y_per = Y_per.iloc[:,-1]
Y_per.reset_index(drop=True, inplace=True)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_per, Y_per, test_size = 0.2, random_state = 42)
```

```python
Loss_function_Total = []
R2_Test_Total = []
R2_Train_Total = []
Accuracy_Test_Total = []
Accuracy_Train_Total = []
```

we know that our data is not regression and we should ude classification but we try regression for seeing the result

## MLPRegressor

```python
mlp = MLPRegressor(hidden_layer_sizes=(20,2), activation='tanh',
                   solver='adam', batch_size=5, learning_rate='constant'
                   ,learning_rate_init=0.001, max_iter=300, shuffle=False,
                   tol=0.00000001, verbose=False, momentum=0.95)
```

```python
mlp.fit(X_train, Y_train)
```

```python
print('Loss : ', mlp.loss_)

train_pred = mlp.predict(X_train)
test_pred = mlp.predict(X_test)
all_pred = mlp.predict(X)

train_pred = (train_pred > 0.5).astype("int32")
test_pred = (test_pred > 0.5).astype("int32")
all_pred = (all_pred > 0.5).astype("int32")

print("Training result - ANN")
print(classification_report(Y_train, train_pred, labels=[1,0]))
print("Testing result - ANN")
print(classification_report(Y_test, test_pred, labels=[1,0])) 
```

```python
loss = mlp.loss_curve_
plt.plot(loss)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()


R2_train = r2_score(Y_train, train_pred)
R_train = np.sqrt(R2_train)

R2_test = r2_score(Y_test, test_pred)
R_test = np.sqrt(R2_test)
# R2 score            .  -1  1     0  1      

print('R2 Train : ', R2_train)
print('R2 Test : ', R2_test)

print('R Train : ', R_train)
print('R Test : ', R_test)
```

```python
Loss_function_Total.append(loss)
R2_Test_Total.append(R2_test)
R2_Train_Total.append(R2_train)
Accuracy_Test_Total.append(accuracy_score(Y_test,test_pred))
Accuracy_Train_Total.append(accuracy_score(Y_train,train_pred))
```

```python
cm = confusion_matrix(Y_test, test_pred)
cm
```

```python
sb.heatmap(cm, center=True)
plt.show()
```

## MLPClassifier

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,100, 100), activation='relu',
                    solver='adam', batch_size=10, learning_rate='adaptive',
                    learning_rate_init=0.001, max_iter=170, shuffle=True,
                    tol=0.00000001, verbose=False, momentum=0.95)
```

```python
mlp.fit(X_train, Y_train)
```

```python
print('Loss : ', mlp.loss_)

train_pred = mlp.predict(X_train)
test_pred = mlp.predict(X_test)
all_pred = mlp.predict(X)

train_pred = (train_pred > 0.5).astype("int32")
test_pred = (test_pred > 0.5).astype("int32")
all_pred = (all_pred > 0.5).astype("int32")

print("Training result - ANN")
print(classification_report(Y_train, train_pred, labels=[1,0]))
print("Testing result - ANN")
print(classification_report(Y_test, test_pred, labels=[1,0])) 
```

```python
loss = mlp.loss_curve_
plt.plot(loss)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()

R2_train = r2_score(Y_train, train_pred)
R_train = np.sqrt(R2_train)

R2_test = r2_score(Y_test, test_pred)
R_test = np.sqrt(R2_test)
print('R2 Train : ', R2_train)
print('R2 Test : ', R2_test)

print('R Train : ', R_train)
print('R Test : ', R_test)
```

```python
Loss_function_Total.append(loss)
R2_Test_Total.append(R2_test)
R2_Train_Total.append(R2_train)
Accuracy_Test_Total.append(accuracy_score(Y_test,test_pred))
Accuracy_Train_Total.append(accuracy_score(Y_train,train_pred))
```

## keras Sequential

```python
model = keras.Sequential(
    [
        layers.Dense(500, input_dim=8, activation="relu"),
        layers.Dense(250, activation="relu"),
        layers.Dense(100, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ]
)
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])    #loss='sparse_categorical_crossentropy', 'binary_crossentropy'
```

```python
model.fit(X_train, Y_train, epochs=85, batch_size=10,verbose = False)
```

```python
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
all_pred = model.predict(X)

train_pred = (train_pred > 0.5).astype("int32")
test_pred = (test_pred > 0.5).astype("int32")
all_pred = (all_pred > 0.5).astype("int32")

print("Training result - ANN")
print(classification_report(Y_train, train_pred, labels=[1,0]))
print("Testing result - ANN")
print(classification_report(Y_test, test_pred, labels=[1,0]))
```

```python
_, train_acc = model.evaluate(X_train, Y_train, verbose=0)
_, test_acc = model.evaluate(X_test, Y_test, verbose=0)
print('Train accuracy: %.3f, Test accuracy: %.3f' % (train_acc, test_acc))
```
